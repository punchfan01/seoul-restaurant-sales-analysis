## 1. 점심 시간대 매출 데이터 가공 (11~17시)

```python
# 점심 시간대 매출 분석
df=pd.read_csv("../data/서울시 상권분석서비스(추정매출-자치구).csv",encoding='cp949')
df=df.loc[:,["기준_년분기_코드","자치구_코드","자치구_코드_명","서비스_업종_코드",
             "서비스_업종_코드_명","시간대_11~14_매출_금액",
             "시간대_14~17_매출_금액"]]
```

```python
# 2023년 3분기
df_20233=df[df.기준_년분기_코드==20233]
df_20233_list=df_20233.values

# 자치구마다 11~14시 매출
dict_20233_sales_11_14={}
for list in df_20233_list:
    key=list[2]
    food_code=int(list[3][2:])
    value=list[5]
    if 100001<=food_code and food_code<=100009:
        dict_20233_sales_11_14[key]=dict_20233_sales_11_14.get(key,0)+value
# print(dict_20233_sales_11_14)

# 자치구마다 14~17시 매출
dict_20233_sales_14_17={}
for list in df_20233_list:
    key=list[2]
    food_code=int(list[3][2:])
    value=list[6]
    if 100001<=food_code and food_code<=100009:
        dict_20233_sales_14_17[key]=dict_20233_sales_14_17.get(key,0)+value
# print(dict_20233_sales_14_17)

# 11~14 + 14~17
dict_20233_sales_11_17={}
for key in dict_20233_sales_11_14:
    dict_20233_sales_11_17[key]=dict_20233_sales_11_14[key]+dict_20233_sales_14_17[key]
# print(dict_20233_sales_11_17)
```

## 2. 유동 인구 데이터 가공

```python
# 유동 인구 수 분석
df=pd.read_csv("../data/서울시 상권분석서비스(길단위인구-자치구).csv",encoding='cp949')
df=df.loc[:,["기준_년분기_코드","자치구_코드","자치구_코드_명","총_유동인구_수"]]

# 2023년 3분기 유동 인구
df_20233=df[df.기준_년분기_코드==20233]
df_20233_list=df_20233.values
dict_20233_people_moving={}
for list in df_20233_list:
    key=list[2]
    value=list[3]
    dict_20233_people_moving[key]=value
# print(dict_20233_people_total)

# 2023년 4분기 유동 인구
df_20234=df[df.기준_년분기_코드==20234]
df_20234_list=df_20234.values
dict_20234_people_moving={}
for list in df_20234_list:
    key=list[2]
    value=list[3]
    dict_20234_people_moving[key]=value
# print(dict_20234_people_total)

# 2024년 1분기 유동 인구
df_20241=df[df.기준_년분기_코드==20241]
df_20241_list=df_20241.values
dict_20241_people_moving={}
for list in df_20241_list:
    key=list[2]
    value=list[3]
    dict_20241_people_moving[key]=value
# print(dict_20241_people_total)

# 2024년 2분기 유동 인구
df_20242=df[df.기준_년분기_코드==20242]
df_20242_list=df_20242.values
dict_20242_people_moving={}
for list in df_20242_list:
    key=list[2]
    value=list[3]
    dict_20242_people_moving[key]=value
# print(dict_20242_people_total)

# 2024년 1분기와 2분기 데이터가 동일
```
